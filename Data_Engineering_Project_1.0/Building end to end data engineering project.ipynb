```python
try:
    # List of queries to alter column types
    alter_queries = [
        "ALTER TABLE accountsdata ALTER COLUMN year_1995 TYPE VARCHAR;",
        "ALTER TABLE accountsdata ALTER COLUMN year_2000 TYPE VARCHAR;",
        "ALTER TABLE accountsdata ALTER COLUMN year_2005 TYPE VARCHAR;",
        "ALTER TABLE accountsdata ALTER COLUMN year_2010 TYPE VARCHAR;",
        "ALTER TABLE accountsdata ALTER COLUMN year_2015 TYPE VARCHAR;"
    ]
    
    # Execute each query
    for query in alter_queries:
        cur.execute(query)
    
    # Commit changes if no errors
    conn.commit()
    print("Data types changed successfully.")

except Exception as e:
    # In case of error, rollback the transaction
    print(f"Error: {e}")
    conn.rollback()  # Rollback the transaction to reset the state
    print("Transaction has been rolled back.")
```
